```python
from quantum_logical.pulsesim.module_build import Module_build, Create_plots
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'ieee'])
```

```python
store_fids = []
lam_power = []
fig, ax = plt.subplots()
for j in range(1,6):
    for i in range(2,7):
        a = Module_build(dim = 2, lam_power = i, mod_count = j)
        store_fids.append(a.results())
        lam_power.append(i)

    # b = Create_plots(x = lam_power, y = store_fids, num_module = j)
    # b.plot_create()
    ax.plot(lam_power, store_fids, marker='o', label = "%s" %(j))
    # ax.set_title("Fidelity vs λ^n")


    

    store_fids.clear()
    lam_power.clear()
# fig.suptitle("Fidelity vs λ^n")
fig.legend(loc = 'center right')
fig.show()
plt.ylabel('Fidelity')
plt.xlabel("n")
plt.title("Fidelity vs $\lambda$")
# plt.legend(title='module count')
# plt.show()
```
