```python
from quantum_logical.pulsesim.module_build import Module_build, Create_plots
import matplotlib.pyplot as plt
```

```python
store_fids = []
lam_power = []
for j in range(1,6):
    for i in range(2,7):
        a = Module_build(dim = 2, lam_power = i, mod_count = j)
        store_fids.append(a.results())
        lam_power.append(i)

    b = Create_plots(x = lam_power, y = store_fids, num_module = j)
    b.plot_create()

    store_fids.clear()
    lam_power.clear()

```

```python
a = Module_build(dim = 2, lam_power = 3, mod_count = 1)
a.results()
```

```python
a = []
x = []
for i in range(5):
    b = Module_build(dim = 2, lam_power = 4, mod_count = i)
    a.append(b.results())
    x.append(i)

plt.plot(x, a, marker = ".")
plt.xticks(x)
plt.xlabel("Module count")
plt.ylabel("Fidelity")
plt.title(f"Fidelity vs Module count for $^{4}$")

plt.savefig("lam_4_plot.svg")

```
