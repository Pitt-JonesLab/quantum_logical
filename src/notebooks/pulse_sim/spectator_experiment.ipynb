```python
from quantum_logical.spectator_testing import spectator
import matplotlib.pyplot as plt
import numpy as np 
import scienceplots
plt.style.use(['ieee'])
plt.rcParams["font.family"] = "Arial"
plt.rcParams['mathtext.rm'] = 'Arial'
```

```python
frequencies_qubit3 = np.linspace(4.000001, 5.9999999, 200)
mag_amp = [[.1, 1.9665551839464883], [.2, 0.4916387959866221], [.3, 0.21872909698996654], [.4, 0.12304347826086957]]
dict1 = {'0':'\u2070','1':'\u00b9','2':'\u00b2',
         '3':'\u00b3','4':'\u2074','5':'\u2075',         
         '6':'\u2076','7':'\u2077','8':'\u2078',
         '9':'\u2079','+':'\u207A','-':'\u207B',
         '=':'\u207C','(':'\u207D',')':'\u207E',
         'a':'\u1d43','b':'\u1d47','c':'\u1D9C',
         'd':'\u1d48','e':'\u1d49','f':'\u1da0',
         'g':'\u1d4d','h':'\u02b0','i':'\u2071',
         'j':'\u02b2','k':'\u1d4f','l':'\u02e1',
         'm':'\u1d50','n':'\u207f','o':'\u1d52',
         'p':'\u1d56','r':'\u02b3','s':'\u02e2',
         't':'\u1d57','u':'\u1d58','v':'\u1d5b',
         'w':'\u02b7','x':'\u02e3','y':'\u02b8',
         'z':'\u1dbb','A':'\u1D2c','B':'\u1D2E',
         'D':'\u1d30','E':'\u1d31','G':'\u1d33',
         'H':'\u1d34','I':'\u1d35','J':'\u1d36',
         'K':'\u1d37','L':'\u1d38','M':'\u1d39',
         'N':'\u1d3a','O':'\u1d3c','P':'\u1d3e',
         'R':'\u1d3f','T':'\u1d40','U':'\u1d41',
         'V':'\u2c7d','W':'\u1d42','q':'\u146b'
         }

normalized_freq = []
for i in range(len(frequencies_qubit3)):
    normalized_freq.append((frequencies_qubit3[i] - frequencies_qubit3[0])/ np.abs(frequencies_qubit3[-1] - frequencies_qubit3[0]))

```

```python
fids2 = []
fids2.clear()
fids4 = []
fids4.clear()
fids6 = []
fids6.clear()
separation2 = []
separation4 = []
separation6 = []
separation2.clear()
separation4.clear()
separation6.clear()
for freq in frequencies_qubit3:
    a = spectator(qubit_3 = freq, lam_power = 2, lam_mag=mag_amp[2][0], amp=mag_amp[2][1])
    fids2.append(a.results())
    separation2.append(a.min_gate_sep())
    b = spectator(qubit_3 = freq, lam_power = 4, lam_mag=mag_amp[2][0], amp=mag_amp[2][1])
    fids4.append(b.results())
    separation4.append(b.min_gate_sep())
    c = spectator(qubit_3 = freq, lam_power = 6, lam_mag=mag_amp[2][0], amp=mag_amp[2][1])
    separation6.append(c.min_gate_sep())
    fids6.append(c.results())

fids2_new = []
```

```python
len(fids2) / 2
sep2_new = []
fids4_new = []
fids6_new = []
sep2_new.clear()
fids4_new.clear()
fids6_new.clear()
```

```python
for i in range(int(len(fids2) / 4)):
    fids2_new.append(fids2[i])
    sep2_new.append(separation2[i])
    fids4_new.append(fids4[i])
    fids6_new.append(fids6[i])

len(fids2_new)
```

```python
fig, axs = plt.subplots()
axs.plot(separation2, fids2, label = '{}'.format(dict1.get('2')))
axs.plot(separation2, fids4, label = '{}'.format(dict1.get('4')))
axs.plot(separation2, fids6, label = '{}'.format(dict1.get('6')))
axs.tick_params(axis="y",direction="in", pad=2)
axs.tick_params(axis="x",direction="in", pad=2)
fig.legend(loc = 'center right')
fig.show()
axs.set_xlabel('Qubit 3 location(Normalized)')
axs.set_ylabel('Fidelity')
axs.tick_params(axis="y",direction="in", pad=2)
axs.tick_params(axis="x",direction="in", pad=2)
image_format = 'svg' # e.g .png, .svg, etc.
image_name = 'spectator_experiment_part_2.svg'
fig.savefig(image_name, format=image_format, dpi=1200)
```
