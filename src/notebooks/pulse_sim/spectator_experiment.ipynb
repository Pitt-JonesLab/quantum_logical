```python
from quantum_logical.spectator_testing import spectator
import matplotlib.pyplot as plt
import numpy as np 
```

```python
frequencies_qubit3 = np.linspace(4.01, 5.99, 200)
```

```python
fids2 = []
fids2.clear()
fids4 = []
fids4.clear()
fids6 = []
fids6.clear()
separation2 = []
separation4 = []
separation6 = []
separation2.clear()
separation4.clear()
separation6.clear()
for freq in frequencies_qubit3:
    a = spectator(qubit_3 = freq, lam_power = 2)
    fids2.append(a.results())
    separation2.append(a.min_gate_sep())
    b = spectator(qubit_3 = freq, lam_power = 4)
    fids4.append(b.results())
    separation4.append(b.min_gate_sep())
    c = spectator(qubit_3 = freq, lam_power = 6)
    separation6.append(c.min_gate_sep())
    fids6.append(c.results())

fids2_new = []
```
