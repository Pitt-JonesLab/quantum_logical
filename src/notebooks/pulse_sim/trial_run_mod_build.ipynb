```python
from quantum_logical.module_build_ import Module_build, Create_plots
import matplotlib.pyplot as plt
```

```python
store_fids = []
lam_power = []
lam_mag = .3
fig, ax = plt.subplots()
for j in range(1,6):
    for i in range(2,7):
        a = Module_build(dim = 2, lam_power = i, mod_count = j, lam_mag = lam_mag, amp = .3)
        store_fids.append(a.results())
        lam_power.append(i)

    # b = Create_plots(x = lam_power, y = store_fids, num_module = j)
    # b.plot_create()
    ax.plot(lam_power, store_fids, marker='o', linestyle = '-', label = "%s" %(j))

    store_fids.clear()
    lam_power.clear()
fig.legend(loc = 'center right')
fig.show()
plt.title(f"Fidelity vs lambda power with varying module count at lambda = {lam_mag}")
plt.xlabel("lambda power")
plt.ylabel("Fidelity")
```
