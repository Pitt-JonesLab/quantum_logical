## Average Gate Fidelity formalism

The goal is to measure the closeness of a completely positive trace-preserving (CPTP) map $\mathcal{E} \in C(\mathcal{X})$ to a target unitary operation $\mathcal{U} \in C(\mathcal{X})$, where $\mathcal{U}[\rho] = U\rho U^\dagger$ for some unitary matrix $U$. This measure of closeness is known as the *Gate Fidelity*.

### Gate Fidelity Definition

The gate fidelity between $\mathcal{E}$ and $\mathcal{U}$ for a quantum state $\rho$ is defined as:

$$
F_{\mathcal{E}, \mathcal{U}}(\rho) = F(\mathcal{E}[\rho], \mathcal{U}[\rho])
$$

where the fidelity function for quantum states, $F(\rho, \sigma)$, is given by:

$$
F(\rho, \sigma) = \left( \text{Tr} \sqrt{\sqrt{\rho} \sigma \sqrt{\rho}} \right)^2
$$

This fidelity function measures the closeness of two quantum states, $\rho$ and $\sigma$.

### Comparison to a Unitary Map

When comparing $\mathcal{E}$ to a unitary map $\mathcal{U}$, the gate fidelity can be expressed in terms of the action of $\mathcal{E}$ and the ideal action of $\mathcal{U}$ on $\rho$:

$$
F_{\mathcal{E}, \mathcal{U}}(\rho) = \left[ \text{Tr} \sqrt{\sqrt{U \rho U^\dagger} \mathcal{E}[\rho] \sqrt{U \rho U^\dagger}} \right]^2
$$

This formulation directly assesses how the operation of $\mathcal{E}$ on a quantum state compares to the desired transformation by $\mathcal{U}$.

### Average Gate Fidelity

The **average gate fidelity** is obtained by averaging $F_{\mathcal{E}, \mathcal{U}}(\rho)$ over all possible quantum states $\rho$, typically utilizing a uniform measure such as the Haar measure for pure states. This averaging process yields:

$$
\bar{F}_{\mathcal{E}, \mathcal{U}} = \int d\psi \, F_{\mathcal{E}, \mathcal{U}}(|\psi\rangle\langle\psi|)
$$

This integral averages the fidelity over all pure states $|\psi\rangle$, providing a comprehensive measure of how closely the channel $\mathcal{E}$ approximates the target unitary operation $\mathcal{U}$ across the entire quantum state space.

## Incoherent error for an iSWAP gate

Consider the Lindblad master equation written as

$$
\dot{\rho} = -i[H, \rho(t)] + \frac{1}{2} \left[ \sum_k \gamma_k \mathcal{D}[L_k] \rho(t) + \sum_k \gamma_{\phi, k} \mathcal{D}[L_k] \rho(t) \right]
$$

where the terms in the Lindblad dissipator are $$\mathcal{D}[L] \rho = 2L \rho L^\dagger - \{L^\dagger L, \rho\}$$

Alternatively the master equation can be written in column vector form as 
$$
\dot{\vec{\rho}} = \mathcal{L} \vec{\rho}
$$
where the matrix $\mathcal{L}$ is the Liouvillian superoperator, 

$$
\mathcal{L} = -i \mathbb{I} \otimes H + i H^T \otimes \mathbb{I} + \frac{1}{2} \left[ \sum_k \gamma_k (2 L_k \otimes L_k - \mathbb{I} \otimes L_k^\dagger L_k - L_k^\dagger L_k \otimes \mathbb{I}) + \sum_k \gamma_{\phi, k} (2 L_k \otimes L_k - \mathbb{I} \otimes L_k^\dagger L_k - L_k^\dagger L_k \otimes \mathbb{I}) \right]
$$, can be shown under vectorization.

The Lindblad equation generates a CPTP map $\mathcal{E}$, therefore we can write the solution to the Lindblad master equation in Kraus operator sum form as 
$$
\rho(t) = \mathcal{E}[\rho(0)] = \sum_i K_i \rho(0) K_i^\dagger
$$

where the Kraus operators $\{K_i\}$ satisfy the completeness relation $\sum_i K_i^\dagger K_i = \mathbb{I}$.

Following vectorization, the solution to the master equation (in column vector form) has the form
$$
\vec{\rho}(t) = e^{\mathcal{L} t} \vec{\rho}(0) = \left[\sum_i K_i^* \otimes K_i \right]\vec{\rho}(0) = \mathcal{S} \vec{\rho}(0)
$$,
where $\mathcal{S}$ is the Lioville superoperator matrix.
