```python
def function(choice, delta_a, delta_b, T):
    if choice == "+":
        func = ((np.sin(T*delta_a + T*delta_b) - 1j*np.cos(T*delta_a + T*delta_b) + 1j) / (delta_a + delta_b))
    elif choice == "-":
        func = ((np.sin(T*delta_a + T*delta_b) + 1j*np.cos(T*delta_a + T*delta_b) - 1j) / (delta_a + delta_b))

    return func
```
